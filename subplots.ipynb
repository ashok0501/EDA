```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
countries=["India","Australia","US","Japan","UK","Germany","Pakistan"]
internetUsage=[95, 95, 90, 82, 88, 95, 40]
```

<h3 style="color: yellow">Sub plots</h3>

```python
fig, ax=plt.subplots(2,2)
ax[0][0].bar(range(1,len(countries)+1),internetUsage)
ax[0][1].plot([1,2,3,4,5],[10,20,20,40,35])
ax[1][1].pie([10,10,12,15,5,25])
ax[1][0].scatter([1,2,3,4,5],[10,10,30,30,50])
plt.show()
```

<h3 style="color: yellow">Annotations</h3>

```python
import matplotlib.pyplot as plt
fig, ax=plt.subplots(2,2)
ax[0][1].plot([1,2,3,4,5],[10,20,20,40,35])
ax[0][1].annotate("Ad 1",xy=(3,20),
                    xytext=(2,30),
                    arrowprops={"arrowstyle":"->"})
ax[0][1].annotate("Ad 1 (removed)",xy=(4,40),
                    xytext=(4,20),
                    arrowprops={"arrowstyle":"->"})
ax[1][1].pie([10,10,12,15,5,25])
ax[1][0].scatter([1,2,3,4,5],[10,10,30,30,50])
plt.show()
```

<h3 style="color: yellow">Row and Column Span</h3>

```python
import matplotlib.pyplot as plt
import random
fig=plt.figure(1,figsize=(5,5))
ax1=plt.subplot2grid((3,2),(0,0),colspan=2)
ax2=plt.subplot2grid((3,2),(1,0))
ax3=plt.subplot2grid((3,2),(1,1),rowspan=2)
ax4=plt.subplot2grid((3,2),(2,0))
ax1.plot([1,2,3,4,5],[10,20,20,40,35])
ax1.annotate("Ad 1",xy=(3,20),
                    xytext=(2,30),
                    arrowprops={"arrowstyle":"->"})
ax1.annotate("Ad 1 (removed)",xy=(4,40),
                    xytext=(4,20),
                    arrowprops={"arrowstyle":"->"})

ax2.scatter([random.randint(1,25) for i in range(100)],[random.randint(10,50) for i in range(100)],c="red")
ax2.scatter([random.randint(1,25) for i in range(100)],[random.randint(10,50) for i in range(100)],c="orange")
ax2.scatter([random.randint(1,25) for i in range(100)],[random.randint(10,50) for i in range(100)],c="black")
ax3.bar([1,2,3,4,5],[10,10,30,30,50])
ax4.boxplot([10,11,14,14,13,11,2,19])
plt.show()
```
