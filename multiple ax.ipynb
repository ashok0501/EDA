```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)
y_inch = np.sin(x) * 10 + 10
fig, ax1 = plt.subplots()
ax1.plot(x, y_inch, 'b-', label='Value in Inches')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Value (inches)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax2 = ax1.twinx()
y_cm = y_inch * 2.54 
ax2.set_ylabel('Value (cm)', color='green')
ax2.tick_params(axis='y', labelcolor='green')
ax2.grid(None) 
fig.tight_layout()
plt.title('Line Graph with Dual Y-Axes')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100) 
y_inch = np.sin(x) * 10 + 10 
fig, ax1 = plt.subplots()
ax1.plot(x, y_inch, 'b-', label='Value in Inches')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Value (inches)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax2 = ax1.twinx()
ax2.spines["right"].set_position(("axes", 1.1))
y_cm = y_inch * 2.54
ax2.set_ylabel('Value (cm)', color='green')
ax2.tick_params(axis='y', labelcolor='green')
ax2.plot(x, y_cm, 'g--', label='Value in CM')
ax1.grid(True)
fig.tight_layout()
plt.title('Line Graph with Side-by-Side Y-Axes')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100) 
y_inch = np.sin(x) * 10 + 10 
fig, ax1 = plt.subplots(figsize=(8, 6))
ax1.plot(x, y_inch, 'b-', label='Value in Inches')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Value (inches)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax2 = ax1.twinx()
ax2.spines["right"].set_position(("axes", 1.2)) 
y_cm = y_inch * 2.54
ax2.set_ylabel('Value (cm)', color='green')
ax2.tick_params(axis='y', labelcolor='green')
ax3 = ax1.twinx()
ax3.spines["right"].set_position(("axes", 1.4)) 
y_mm = y_inch * 25.4
ax3.set_ylabel('Value (mm)', color='purple')
ax3.tick_params(axis='y', labelcolor='purple')
ax4 = ax1.twinx()
ax4.spines["right"].set_position(("axes", 1.6)) 
y_m = y_inch * 0.0254
ax4.set_ylabel('Value (m)', color='orange')
ax4.tick_params(axis='y', labelcolor='orange')
plt.title('Line Graph with Four Y-Axes (Inches, cm, mm, m)')
ax1.grid(True)
fig.tight_layout()
plt.show()
```
